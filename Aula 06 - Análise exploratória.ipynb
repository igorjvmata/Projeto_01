```python
#Criando o DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Visualizar a quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a receita total?
df[ "Valor Venda"].sum()
```

```python
#Qual o custo total?
#Cria coluna de Custo
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
#Qual o custo total?
round(df["Custo"].sum(),2)
```

```python
#De posse da receita e custo, podemos encontrar o lucro total
#Criando a coluna de Lucro
df["Lucro"] = df ["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
#Total do lucro
round(df["Lucro"].sum(),2)
```

```python
#Criando uma coluna com o total de dias para enviar o produto
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas
```

```python
df.shape
```

**Buscando encontrar a média do tempo de envio por marca**

```python
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
df["Tempo Envio"].dtype
```

```python
#Média do tempo de envio por marca
df.groupby("Marca")["Tempo Envio"].mean()
```

Missing Values

```python
#Verificando se existem dados faltantes
df.isnull().sum()
```

Buscando o lucro por ano e por marca

```python
#Agrupar por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Gráfico Lucro x Ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
#Lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Gráfico de Lucro x Mês do ano de 2009
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Gráfico de Lucro x Marca do ano de 2009
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Lucro")
plt.ylabel("Marca");
```

```python
#Gráfico de Lucro x Classe do ano de 2009
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Marca")
plt.xticks(rotation='horizontal');
```
